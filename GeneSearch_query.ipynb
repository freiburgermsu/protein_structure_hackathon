# Load the FASTA data

```python
with open("D:/Users/Andrew Freiburger/Documents/Argonne/Encoders/EcoliMG1655/EcoliMG1655.md5.fa", "r") as fa:
    fastaFile = fa.readlines()

indexed_fasta = {"sequences": [], "params": {"eValue": 1e-100, "maxHits": 1}}
for line in fastaFile:
    line = line.replace("\n", "")
    if ">" in line:
        # if any([line in seq["id"] for seq in indexed_fasta["sequences"]]):
        #     indexed_fasta["sequences"]
        indexed_fasta["sequences"].append({"id": line[1:], "data": ""})
    else:
        indexed_fasta["sequences"][-1]["data"] += line
        
print(indexed_fasta["sequences"][-1])
```

# Bin the sequences

```python
queries = [{"sequences": [], "params": {"eValue": 1e-100, "maxHits": 1}}]
for seq in indexed_fasta["sequences"]:
    if len(queries[-1]["sequences"]) < 5:
        queries[-1]["sequences"].append(seq)
    else:
        queries.append({"sequences": [], "params": {"eValue": 1e-100, "maxHits": 1}})
print(len(queries))
```

# Submit the jobs

```python
from collections import OrderedDict
from pprint import pprint
from requests import post
from json import loads

queries_hits2 = []
for query in queries[1:]:
    print(query)
    query_hits = post("https://genesearch2.kbase.us/api/v1/namespaces/uniref100/jobs", json=query, verify=False)
    print(query_hits.text)
    # ordered_dict = OrderedDict(loads(query_hits.text))
    # pprint(ordered_dict['alignments'][0])
    queries_hits2.append(query_hits.text)
    # break
```
