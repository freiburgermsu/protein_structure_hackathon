```python
from collections import Counter
counts = Counter(sprot["id"])
md5_uniprot_exchange = {}
for key, score in tmscores.items():
    ID = re.search(r"(?<=AF-)(\w+)(?=-F1)", key).group()
    if ID not in counts:
        print("missing:\t", ID)
        continue
    if counts[ID] == 1:
        md5 = sprot[sprot["id"] == ID]["md5"].values[0]
        # print(md5, key)
        md5_uniprot_exchange[md5] = key        
        pass
    else:
        print(f"ID with {counts[ID]} counts:", ID)
        break
# display(uniprot_mapped)
# Genes_df
```

```python
sims["tm"] = [None]*len(sims["prot1"])
for index, row in sims.iterrows():
    if row["prot1"] not in md5_uniprot_exchange:
        print(row["prot1"])
        break
        continue
    uniprot1 = md5_uniprot_exchange[row["prot1"]]
    if row["prot2"] not in md5_uniprot_exchange:
        continue
    uniprot2 = md5_uniprot_exchange[row["prot2"]]
    
    score1, score2 = tmscores[uniprot1], tmscores[uniprot2]
    print(score2-score1)
    sims.at[index, "tm"] = score1-score2
```

```python
import re

uniprot_ids = set()
for key, score in tmscores.items():
    uniprot_ids.add(re.search(r"(?<=AF-)(\w+)(?=-F1)", key).group())

uniprot_id_string = ""
row_count = 0
for ID in uniprot_ids:
    if row_count%10 != 0:
        uniprot_id_string += ID+"\t"
    else:
        uniprot_id_string += ID+"\n"
    row_count += 1
    
with open("D:/Users/Andrew Freiburger/Documents/Argonne/Encoders/tmscore_uniprot_ids.txt", "w") as out:
    out.write(uniprot_id_string)
```

```python
# load the mappings
from pandas import read_table, read_excel
from json import dump
ID_mappings = read_table("D:/Users/Andrew Freiburger/Documents/Argonne/Encoders/uniref100_TMscore_IDs.tsv")
uniprot_to_uniref100 = {}
for index, row in ID_mappings.iterrows():
    uniprot_to_uniref100[row["From"]] = re.search(r"(?<=\_)(\w+)", row["Cluster ID"]).group()
# print(uniprot_to_uniref100)

with open("uniprot_uniref_conversion.json", "w") as out:
    dump(uniprot_to_uniref100, out, indent=3)
```

```python
# translate the mappings
from json import load, dump
with open("uniprot_uniref_conversion.json", "r") as data:
    uniprot_to_uniref100 = load(data)
    
missing_IDs = []
translated_tmscores = {}
for key, score in tmscores.items():
    ID = re.search(r"(?<=AF-)(\w+)(?=-F1)", key).group()
    if ID not in uniprot_to_uniref100:
        missing_IDs.append(ID)
    else:
        translated_tmscores[uniprot_to_uniref100[ID]] = score

# export the translated mappings
with open("D:/Users/Andrew Freiburger/Documents/Argonne/Encoders/uniref100_tmscores.json", "w") as out:
    dump(translated_tmscores, out, indent=3)
```

```python
with open("D:/Users/Andrew Freiburger/Documents/Argonne/Encoders/EcoliMG1655/EcoliMG1655.md5.fa", "r") as fa:
    fastaFile = fa.readlines()

indexed_fasta = {"sequences": [], "params": {"eValue": 1e-100, "maxHits": 1}}
for line in fastaFile:
    line = line.replace("\n", "")
    if ">" in line:
        # if any([line in seq["id"] for seq in indexed_fasta["sequences"]]):
        #     indexed_fasta["sequences"]
        indexed_fasta["sequences"].append({"id": line, "data": ""})
    else:
        indexed_fasta["sequences"][-1]["data"] += line
        
print(indexed_fasta["sequences"][-1])
```

```python
queries = [{"sequences": [], "params": {"eValue": 1e-100, "maxHits": 1}}]
for seq in indexed_fasta["sequences"]:
    if len(queries[-1]["sequences"]) < 5:
        queries[-1]["sequences"].append(seq)
    else:
        queries.append({"sequences": [], "params": {"eValue": 1e-100, "maxHits": 1}})
print(len(queries))
```
