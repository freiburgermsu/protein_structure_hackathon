### missing uniprot ID's in Sprot

```python
from collections import Counter
counts = Counter(sprot["id"])
md5_uniprot_exchange = {}
for key, score in tmscores.items():
    ID = re.search(r"(?<=AF-)(\w+)(?=-F1)", key).group()
    if ID not in counts:
        print("missing:\t", ID)
        continue
    if counts[ID] == 1:
        md5 = sprot[sprot["id"] == ID]["md5"].values[0]
        # print(md5, key)
        md5_uniprot_exchange[md5] = key        
        pass
    else:
        print(f"ID with {counts[ID]} counts:", ID)
        break
# display(uniprot_mapped)
# Genes_df
```

```python
iteration = 0
for k, v in md5_uniprot_exchange.items():
    print(k, v)
    iteration += 1
    if iteration == 5:
        break
```

```python
sims["tm"] = [None]*len(sims["prot1"])
for index, row in sims.iterrows():
    if row["prot1"] not in md5_uniprot_exchange:
        print(row["prot1"])
        break
        continue
    uniprot1 = md5_uniprot_exchange[row["prot1"]]
    if row["prot2"] not in md5_uniprot_exchange:
        continue
    uniprot2 = md5_uniprot_exchange[row["prot2"]]
    
    score1, score2 = tmscores[uniprot1], tmscores[uniprot2]
    print(score2-score1)
    sims.at[index, "tm"] = score1-score2
```
