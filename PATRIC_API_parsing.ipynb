### sequence object parsing

```python
seqs = {}
for gene in ary:
    if "na_sequence_md5" not in gene:
        continue
    if gene["sequence_id"] in seqs:
        seqs[gene["sequence_id"]].append(gene["na_sequence_md5"])
    else:
        seqs[gene["sequence_id"]] = [gene["na_sequence_md5"]]
```

```python
for seqID, md5s in seqs.items():
    for md5 in md5s:
        print(seqID, md5)
        sequence = get(f"https://bv-brc.org/api/genome_sequence/?eq(sequence_id,{seqID})&eq(sequence_md5,{md5})")
        try:
            with open(f"Ecoli/sequences/NA_MD5s/{md5}.json", 'w') as out:
                json.dump(sequence.json(), out, indent=3)
        except:
            with open(f"Ecoli/sequences/NA_MD5s/{md5}.txt", 'w') as out:
                out.write(sequence.text)
```
