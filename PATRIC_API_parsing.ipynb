### Process the features into a desirable table

```python
from pprint import pprint
pprint(ary[0])
```

```python
from pandas import DataFrame

patricIDs, protein, genomeID, genomeName, pgfamID, plfamID, aa_sequence_md5, na_sequence_md5 = [], [], [], [], [], [], [], []
for gene in ary:
    patricIDs.append(gene.get("patric_id", " "))
    protein.append(gene.get("product", " "))
    genomeID.append(gene.get("genome_id", " "))
    genomeName.append(gene.get("genome_name", " "))
    pgfamID.append(gene.get("pgfam_id", " "))
    plfamID.append(gene.get("plfam_id", " "))
    aa_sequence_md5.append(gene.get("aa_sequence_md5", " "))
    na_sequence_md5.append(gene.get("na_sequence_md5", " "))
    
features_table = DataFrame({"PATRIC ID's":patricIDs, "Protein name":protein, "Protein MD5":aa_sequence_md5, "Genome ID": genomeID, 
                            "Genome name": genomeName, "Gene MD5":na_sequence_md5, "pgfam ID": pgfamID, "plfam ID": plfamID})
display(features_table)
features_table.to_csv("Ecoli/features_table.csv")
```

### sequence object parsing

```python
seqs = {}
for gene in ary:
    if "na_sequence_md5" not in gene:
        continue
    if gene["sequence_id"] in seqs:
        seqs[gene["sequence_id"]].append(gene["na_sequence_md5"])
    else:
        seqs[gene["sequence_id"]] = [gene["na_sequence_md5"]]
```

```python
for seqID, md5s in seqs.items():
    for md5 in md5s:
        print(seqID, md5)
        sequence = get(f"https://bv-brc.org/api/genome_sequence/?eq(sequence_id,{seqID})&eq(sequence_md5,{md5})")
        try:
            with open(f"Ecoli/sequences/NA_MD5s/{md5}.json", 'w') as out:
                json.dump(sequence.json(), out, indent=3)
        except:
            with open(f"Ecoli/sequences/NA_MD5s/{md5}.txt", 'w') as out:
                out.write(sequence.text)
```
